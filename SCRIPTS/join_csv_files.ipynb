```python
import pandas as pd

# File paths (need to change directories)
housing_file = "/Users/emilymoore/Downloads/DS 4002 Project 2/caar_market_reports/caar_data.csv"
sentiment_file = "/Users/emilymoore/Downloads/DS 4002 Project 2/meeting_mins/housing_uncertainty_with_score.csv"
output_file = "/Users/emilymoore/Downloads/DS 4002 Project 2/combined_housing_sentiment.csv"

# Load the datasets
housing_df = pd.read_csv(housing_file)
sentiment_df = pd.read_csv(sentiment_file)

# Remove file name columns if they exist
housing_df = housing_df.drop(columns=["file_name"], errors="ignore")
sentiment_df = sentiment_df.drop(columns=["file_name"], errors="ignore")

# Combine datasets (assumes rows correspond month-by-month)
df = pd.concat([housing_df, sentiment_df], axis=1)

# Convert median price to numeric
df["median_price"] = pd.to_numeric(df["median_price"], errors="coerce")

# Calculate median price change percentage
df["median_price_change_pct"] = df["median_price"].pct_change() * 100

# Calculate sentiment score
df["sentiment_score"] = (
    (df["housing_mentions"] - df["uncertainty_mentions"]) /
    df["minutes_word_length"]
)

# Remove and rearrange columns
df = df.drop(columns = ['start_time', 'end_time', 'housing_uncertainty_sentences'], errors = "ignore")
new_column_order = ['month_year', 'median_price', 'median_price_change_pct', 'sentiment_score', 'sales_volume', 'uncertainty_mentions', 
                    'uncertainty_score', 'minutes_word_length']
df = df[new_column_order]

# Save combined dataset
df.to_csv(output_file, index=False)

print("Combined dataset saved to:", output_file)
```
