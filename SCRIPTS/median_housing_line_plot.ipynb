```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt

# Change the path to match the proper path on your computer
df = pd.read_csv("/Users/emilymoore/Downloads/DS 4002 Project 2/combined_housing_sentiment.csv")
df['month_year'] = pd.to_datetime(df['month_year']) # Change the format of month_year column to match "YYYY-MM"
df = df.sort_values('month_year') # Sort the dates in chronological order

plt.figure(figsize=(10,5)) # Set size for the plot
plt.plot(df['month_year'], df['median_price'], marker='o', label='Median Price') # Columns used for the plot, marking each point with a dot
plt.title("Median Housing Price Over Time") # Proper title for the plot
plt.xlabel("Month") # Label the x-axis
plt.ylabel("Median Price ($)") # Label the y-axis
plt.grid(True) # Added grid to help visualize data points
plt.xticks(rotation=45) 
plt.legend() # Added legend in right-hand corner by default
plt.savefig("median_price_over_time.png", dpi=300, bbox_inches='tight') # Save the figure as a png file to Jupyter Notebook
plt.show() # Show the figure
```
